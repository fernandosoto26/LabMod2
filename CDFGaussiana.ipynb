```python
#---------------------------------------------------------
#---Es para CDF (cumulative distribution function)
#---de una gaussiana
#---------------------------------------------------------
#-----TODAS LAS LIBRERIAS QUE PODRIAN SER DE UTILIDAD-----
import random #proporciona un generador de números aleatorios
import math #Provee acceso a funciones matematicas estandar.
from matplotlib import pyplot as plt #Para visualizaciones y gráficos. Nos va a permitir realizar los gráficos de las distintas distribuciones de datos.
import numpy as np #proporciona algunas funciones estadísticas que podemos aplicar fácilmente sobre los arrays de Numpy.
import seaborn as sns # importando seaborn; Esta librería es un complemento ideal de matplotlib para realizar gráficos estadísticos.
import pandas as pd #librería para análisis de datos y financieros. Posee funciones de estadística descriptiva sobre nuestros datos y
from mpl_toolkits.mplot3d import axes3d #permite hacer graficas en 3D, suministrando ejes a nuestro objeto

# parametros esteticos de seaborn
sns.set_palette("dark", desat=.6) #establecerá el ciclo de color predeterminado para todas las graficas
                #"deep" variaciones del tema default(existen 6), desat es para desaturar cada color  
sns.set_context(rc={"figure.figsize": (8, 4)}) #Establezca los parámetros del plotting. Afecta el tamaño de las etiquetas, 
                                               #líneas y otros elementos de la trama, pero no el estilo general.
mu = 70 # media
sigma = 2 # y desvio estandar
N=1000 
s = np.random.normal(mu, sigma, N) #Dibuja muestras aleatorias de una distribución normal (gaussiana). (loc=0.0, scale=1.0, size=None)
# histograma de distribución normal.
cuenta, cajas, ignorar = plt.hist(s, 100, density=True) #Es para formar las barras de la funcion normal aleatoria,
                                                        #el 100 corresponde a la cantidad de barras
normal = plt.plot(cajas, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (cajas - mu)**2 / (2 * sigma**2) ),linewidth=2, color='r') 
#Corresponde a la funcion P(x|mu,sigma)= (1/(sigma raiz(2pi)))* EXP(-(-(x-mu)^2)/(2 sigma^2))
plt.show() #Graficar en diferentes graficas (son 2)
#Para generar el CDF (cumulative distribution function)
X2 = np.sort(s) #Devuelve una copia ordenada de un arreglo, es decir, los valores de la gaussiana se ordenaron 
F2 = np.array(range(N))/float(N) #Crea un arreglo.
plt.plot(X2, F2,linewidth=5, color='r') #graficado de CDF
plt.show()
```
