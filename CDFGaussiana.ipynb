```python
#---------------------------------------------------------
#---Es para CDF (cumulative distribution function)
#---de una gaussiana
#---------------------------------------------------------
#-----TODAS LAS LIBRERIAS QUE PODRIAN SER DE UTILIDAD-----
import random #proporciona un generador de números aleatorios
import math #Provee acceso a funciones matematicas estandar.
from matplotlib import pyplot as plt #Para visualizaciones y gráficos. 
import numpy as np #proporciona algunas funciones estadísticas
import seaborn as sns # para realizar gráficos estadísticos.
from mpl_toolkits.mplot3d import axes3d #permite hacer graficas en 3D


sns.set_palette("dark", desat=.6) #establecerá el ciclo de color predeterminado para todas las graficas       
sns.set_context(rc={"figure.figsize": (8, 4)}) #Establezca los parámetros del plotting. Afecta el tamaño de las etiquetas, 
                                               
mu = 70 # media
sigma = 2 #desvio estandar
N=1000  #valores
s = np.random.normal(mu, sigma, N) #Dibuja muestras aleatorias de una distribución normal (gaussiana). (loc=0.0, scale=1.0, size=None)
# histograma de distribución normal.
a,  b , c= plt.hist(s, 100, density=True) #Es para formar las barras de la funcion normal aleatoria,
#cuenta,  cajas , ignorar
#el 100 corresponde a la cantidad de barras
normal = plt.plot(b, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (b - mu)**2 / (2 * sigma**2) ),linewidth=2, color='r') 
#Corresponde a la funcion P(x|mu,sigma)= (1/(sigma raiz(2pi)))* EXP(-(-(x-mu)^2)/(2 sigma^2))
plt.show() #Graficar en diferentes graficas (son 2)
#Para generar el CDF (cumulative distribution function)
X2 = np.sort(s) #Devuelve una copia ordenada de un arreglo, es decir, los valores de la gaussiana se ordenaron 
F2 = np.array(range(N))/float(N) #Crea un arreglo.
plt.plot(X2, F2,linewidth=5, color='r') #graficado de CDF
plt.show()
```
