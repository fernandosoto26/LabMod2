```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

N = 500 # puntos
t = 20  #tamaño caja
dat = t*np.random.rand(N,2) #"datos"
ale = t*np.random.rand(N,2) #aleatorio

#distancia entre puntos
def  distancia(x1,x2,y1,y2):
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
    return d

#matriz simetrica
dtot = N*(N-1)/2
j = 0
dold = 0
dmax = 0
disdat = []
disale = []

for i in range(N):
    for j in range(i+1,N):
        d_d = distancia(dat[i][0],dat[j][0],dat[i][1],dat[j][1])
        d_r = distancia(ale[i][0],ale[j][0],ale[i][1],ale[j][1])
        disdat.append(d_d)
        disale.append(d_r)
        
#datos
plt.plot(dat,'o')
plt.show()
#hist. datos
datn, datbin ,m = plt.hist(disdat,bins=100, color='green')
plt.show()
#hist aleatorio
alen, alebin, n = plt.hist(disale,bins=100, color='yellow' )
plt.show()
#hist 2d datos-aleatorio
plt.hist2d(disdat, disale , bins=100)
cb = plt.colorbar()
```
