```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
%matplotlib inline 

#Formando el anillo parte superior 
theta = (np.random.random(1000))*(-1)*np.pi
r1 = np.random.uniform(500.0,1000.,1000)
r = np.sqrt(r1)
xs = r*np.cos(theta) + 50. #x inferior
ys = r*np.sin(theta) + 50. #y inferior 

#Formando el rectangulo lado l
l = 200
recx = np.random.uniform(0.,100.,1000)
recy = np.random.uniform(0.,100.,1000)

#grafica puntos disco y rectangulo
plt.plot(recx,recy,'.')
plt.plot(xs,ys,'.',color='green')
plt.show()

#distancia entre puntos 2D
def distancia(x1,x2,y1,y2):
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
    return d
#guardando dist.
dato = []
ran = [] 
w1=1 #Funcion de peso igual a 1
for i in range(len(w1*xs)):
    for j in range(i+1,len(w1*xs)):
        dd = distancia(xs[i],xs[j],ys[i],ys[j])
        dr = distancia(recx[i],recx[j],recy[i],recy[j])
        dato.append(dd)
        ran.append(dr)
nbin = np.linspace(0.,np.sqrt(2.)*l,20)
#print nbin        
#histograma datos DD
daton, datobin,m = plt.hist(dato,bins=nbin)
plt.title('Histograma DD')
plt.show()
#histograma aleatorios RR
randn, randbin, p = plt.hist(ran,bins=nbin, color='green')
plt.title('Histograma RR')
plt.show()

dr = []
for i in range(len(w1*xs)):
    for j in range(len(w1*xs)):
        ddr = distancia(xs[i],recx[j],ys[i],recy[j])
        dr.append(ddr)
    
#Histograma DR    
drn, drbin, q = plt.hist(dr,bins=nbin, color='yellow')
plt.title('Histograma DR')
plt.show()

#Para formar estimadores
DD=daton
RR=randn
DR=drn


#Fun. de correlacion
#Hamilton
plt.plot(datobin[1:], (DD*RR)/(DR)**2  ,'r')
plt.title('Hamilton')
plt.axis([0, 100, 0, 1])
plt.show()

#Landy 
plt.plot(datobin[1:],1.+ DD/RR -2*DR/RR,'g')
plt.title('Landy')
plt.show()

#Peebels
plt.plot(datobin[1:], DD/RR - 1. ,'b')
plt.title('Peebels')
plt.show()

#Davis
plt.plot(datobin[1:], (DD)/(DR) -1.  ,'y')
plt.title('Davis')
plt.show()
```
