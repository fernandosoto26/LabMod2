```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
import scipy.fftpack
%matplotlib inline

x = np.random.random(1024) #aleatorios

def DFT_slow(x):
    """Compute the discrete Fourier Transform of the 1D array x"""
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(-2j * np.pi * k * n / N)
    return np.dot(M, x)

N = 5000
data = np.random.rand(N)

t = np.linspace(-13,13,5000)
r = np.exp(-t)

plt.plot(t,r)
plt.show()



F = DFT_slow(r)
plt.plot(F)
plt.show()

%timeit DFT_slow(x)
%timeit np.fft.fft(x)
```
